```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('House Price India.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

### Univariate 

```python
plt.figure(figsize=(10, 10))
sns.histplot(x=df['living area'], kde=True)
```

```python
df['number of floors'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(x=df['number of floors'].value_counts(), autopct='%.2f%%')
plt.legend(df['number of floors'].value_counts().index)
plt.show()
```

```python
sns.boxplot(x=df['Area of the house(excluding basement)'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(df['condition of the house'].value_counts(), autopct='%.0f%%')
plt.legend(df['condition of the house'].value_counts().index)
plt.show()
```

### Bivariate

```python
sns.scatterplot(y=df['Price'], x=df['Area of the basement'])
```

```python
sns.scatterplot(x=df['living area'], y=df['Price'])
```

### Multivariate

```python
plt.figure(figsize=(13, 13))
sns.heatmap(df.corr(), annot=True, fmt='.1f')
plt.show()
```
